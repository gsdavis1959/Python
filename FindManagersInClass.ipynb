```python
import pandas as pd

# Load the Excel file
file_path = '/content/classes.xlsx'  # Replace with your file path
df = pd.read_excel(file_path)


df
```

```python
# Create an empty list to store the matches
matches = []

# Iterate over each unique class in the DataFrame
for class_name in df['Class'].unique():
    # Filter DataFrame for the current class
    class_df = df[df['Class'] == class_name]

    # Iterate over each participant in the current class
    for participant in class_df['Participant'].unique():
        # Find the manager of the participant within the current class
        manager_df = class_df[class_df['Manager'] == participant]['Participant'] # Get the DataFrame of managers for the current participant
        manager = manager_df.iloc[0] if not manager_df.empty else None # Check if the DataFrame is empty before accessing the first element

        matches.append({'Class': class_name,
                       'Participant': manager,
                       'Manager in class': participant})

# Create a DataFrame from the matches list
matches_df = pd.DataFrame(matches)

# Print the DataFrame
print(matches_df)
```
