```python
import pandas as pd

# Load the dataset.
df = pd.read_csv('https://raw.githubusercontent.com/gsdavis1959/Data/master/ProductReviews.csv')
df.head()
```

```python

# Create a user-item matrix.
user_item_matrix = df.pivot_table(index="UserId", columns="ProductId", values="HelpfulnessNumerator")

# Calculate the similarity between users.
similarity = user_item_matrix.corr()
similarity.head()

```

```python
# Create a list of recommendations for each user.
recommendations = []
for UserId in similarity.index:
    recommendations.append(list(similarity.loc[UserId].sort_values().tail(10).index))
rec_df = pd.DataFrame(recommendations)
rec_df.head()
```

```python


# Display the recommendations to the user.
for UserId, recommendation in zip(df["UserId"], recommendations):
    print("Recommendations for user {}:".format(UserId))
    for ProductId in recommendation:
        print(ProductId)
```

```python
user_item_matrix = user_item_matrix.fillna(0)
```

```python
similarity_matrix = 1 - pairwise_distances(user_item_matrix, metric='correlation')
```

```python
similarity_df = pd.DataFrame(similarity_matrix, index=user_item_matrix.index, columns=user_item_matrix.index)
```

```python
similarity_df
```

```python
# Assume you want to find the top N similar users for each user
N = 5

# Iterate through each user and find the top N similar users
for user in user_item_matrix.index:
    similar_users = similarity_df[user].sort_values(ascending=False)[1:N+1]  # Exclude the user's own similarity
    for similar_user, similarity_score in similar_users.items():
        df.loc[df['UserId'] == user, f'Similarity_{similar_user}'] = similarity_score
```

```python
import pandas as pd
# Load the dataset.
df = pd.read_csv('https://raw.githubusercontent.com/gsdavis1959/Data/master/ProductReviews.csv')
df.head()
# Sample DataFrame

df = df.rename(columns={'Score': 'Rating', 'ProductId': 'Title'})
df
```

```python
# Pivot the DataFrame to create a user-title matrix
user_title_matrix = df.pivot(index='UserId', columns='Title', values='Rating')
user_title_matrix = user_title_matrix.fillna(0)
user_title_matrix.head()
```

```python



# Calculate the similarity matrix using pairwise distances (e.g., Pearson correlation)
similarity_matrix = 1 - pairwise_distances(user_title_matrix.T, metric='correlation')
similarity_df = pd.DataFrame(similarity_matrix, index=user_title_matrix.columns, columns=user_title_matrix.columns)

# Define the number of top similar titles to consider for each user
N = 5

# Create an empty DataFrame to store the top N similar titles for each user
top_similar_titles_df = pd.DataFrame(columns=['UserId', f'Top{N}SimilarTitles'])

# Iterate through each user
for user in user_title_matrix.index:
    # Get the user's rated titles
    rated_titles = user_title_matrix.loc[user, user_title_matrix.loc[user] != 0].index

    # Initialize a dictionary to store similarity scores for each title
    title_similarity_scores = {}

    # Iterate through the user's rated titles
    for title in rated_titles:
        # Get the similarity scores for the title
        title_similarity = similarity_df[title].sort_values(ascending=False)

        # Exclude the title itself from the list of similar titles
        similar_titles = title_similarity.drop(title).index

        # Add similarity scores to the dictionary
        title_similarity_scores[title] = similar_titles[:N].tolist()

    # Append the results to the new DataFrame
    top_similar_titles_df = top_similar_titles_df.append({
        'UserId': user,
        f'Top{N}SimilarTitles': title_similarity_scores
    }, ignore_index=True)

# Display the DataFrame with top N similar titles for each user
print(top_similar_titles_df)
```
