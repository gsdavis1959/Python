```python
import pandas as pd

# Create the original DataFrame
data = {
    'ID': [1, 1, 1, 2, 2, 2, 3, 3, 3],
    'Course': ['A', 'B', 'C', 'C', 'B', 'A', 'B', 'A', 'C'],
    'Sequence': [1, 2, 3, 1, 2, 3, 1, 2, 3]
}

df = pd.DataFrame(data)

# Pivot the DataFrame to create the "Before" and "After" columns
pivot_df = df.pivot(index='ID', columns='Sequence', values='Course').reset_index()

# Rename the columns to "Before" and "After"
pivot_df.columns = ['ID', 'Before1', 'Before2', 'Before3']

# Create a new DataFrame with "Before" and "After" columns
result_df = pd.DataFrame()
result_df['Before'] = pivot_df[['Before1', 'Before2']].apply(lambda x: ' '.join(x), axis=1)
result_df['After'] = pivot_df[['Before2', 'Before3']].apply(lambda x: ' '.join(x), axis=1)

# Display the resulting DataFrame
print(result_df)

```

```python
import pandas as pd

# Create the original DataFrame
data = {
    'ID': [1, 1, 1, 2, 2, 2, 3, 3, 3],
    'Course': ['A', 'B', 'C', 'C', 'B', 'A', 'B', 'A', 'C'],
    'Sequence': [1, 2, 3, 1, 2, 3, 1, 2, 3]
}

df = pd.DataFrame(data)
df


```

```python
# Sort the DataFrame by 'ID' and 'Sequence'
df.sort_values(by=['ID', 'Course','Sequence'], inplace=True)

# Create a new DataFrame for 'Before' and 'After'
result_df = pd.DataFrame()

# Extract 'Before' and 'After' values based on sequence
result_df['Before'] = df['Course'].shift()
result_df['After'] = df['Course'].shift(1)

# Drop rows with NaN values
result_df.dropna(inplace=True)

# Reset the index
result_df.reset_index(drop=True, inplace=True)

# Display the resulting DataFrame
print(result_df)
```
