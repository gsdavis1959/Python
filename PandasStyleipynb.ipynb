```python
import pandas as pd
data = {'product_name': ['pencil', 'eraser', 'pen', 'desk', 'laptop','laptop','pen','pencil','pen','eraser','gift','pencil',''],
        'price': [2, 1, 1, 50, 678,500,2,5,3,-9,-2,-1,10], 
        'owner': ['Chi','Joe','Dan','Lisa','Rose','Chi','Dan','Lisa','Joe','Rose','Jen','Dan','Mike']}

df = pd.DataFrame(data)
```

```python
df.style.hide_columns('price')
```

```python
df.style.highlight_null(null_color="yellow")
```

```python
df.groupby('owner').agg({'price':'sum'}).reset_index().style.highlight_max(color = 'green') 
```

```python
df.groupby('owner').agg({'price':'sum'}).reset_index().style.highlight_min(color = 'red')
```

```python
import numpy as np
np.random.seed(100)
df3 = pd.DataFrame({'A': np.linspace(1, 8, 10)})
df3 = pd.concat([df3, pd.DataFrame(np.random.randn(8, 4), columns=list('BCDE'))],
               axis=1)
df3.iloc[3, 1] = np.nan
df3.iloc[1, 1] = np.nan
```

```python
import seaborn as sns
cm = sns.light_palette("pink", as_cmap=True)
df3.style.background_gradient(cmap=cm) 
```

```python
df3.style.bar(subset=['A', 'B'], align="zero", color=['pink', 'lightgreen'])
```
