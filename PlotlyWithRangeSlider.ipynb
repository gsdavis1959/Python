```


import pandas as pd
import plotly.graph_objects as go

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/gsdavis1959/Data/master/WorldPopulation.csv')
# Convert 'Year' to datetime and '80+' to numeric
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df['80+'] = pd.to_numeric(df['80+'], errors='coerce')
# Filter rows where 'Year' is between 2022 and 2100
df = df[(df['Year'].dt.year >= 2022) & (df['Year'].dt.year <= 2100)]
```

```
# Subset of countries
countries = ['Burundi', 'Kenya', 'Rwanda', 'Uganda', 'Japan', 'Italy', 'Canada']

# Create the initial plot
fig = go.Figure()
# Add a line for each country
for country in countries:
    country_data = df[df['Region, subregion, country or area *'] == country]
    fig.add_trace(go.Scatter(x=country_data['Year'], y=country_data['80+'], name=country))
```

```
# Add a range slider and dropdown menu
fig.update_layout(
    xaxis=dict(
        rangeslider=dict(visible=True),
        type='date',
        range=['2022', '2100']
    ),
    title='Population Index over the Years for Age Group 80+',
    xaxis_title='Year',
    yaxis_title='Population Index (%)',
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=["visible", [True]*len(countries)],
                    label="All",
                    method="restyle"
                )] +
                [dict(
                    args=["visible", [country == c for c in countries]],
                    label=country,
                    method="restyle"
                ) for country in countries]
            ),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top"
        ),
    ]
)
# Show the plot
fig.show()
```
