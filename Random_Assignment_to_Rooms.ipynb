```python
import pandas as pd
import random

# Creating a fake list of 100 people with id and names
people = [{'id': i, 'name': f'Person_{i}'} for i in range(1, 101)]

# Defining the rooms and conditions
rooms = ['Room 1', 'Room 2', 'Room 3']
room_capacities = {'Room 1': 25, 'Room 2': 30, 'Room 3': 45}
conditions = ['Condition A', 'Condition B', 'Condition C', 'Condition D']

# Function to assign people to rooms based on capacity and assign conditions
def assign_rooms_properly(people, capacities):
    final_assignments = []

    for condition in conditions:
        # Shuffle the people for each condition to ensure randomness
        random.shuffle(people)

        # Assign people to rooms based on the given capacities
        room_assignments = {
            'Room 1': people[:capacities['Room 1']],
            'Room 2': people[capacities['Room 1']:capacities['Room 1'] + capacities['Room 2']],
            'Room 3': people[capacities['Room 1'] + capacities['Room 2']:]
        }

        # Create assignment for each person with room and condition
        for room, assigned_people in room_assignments.items():
            for person in assigned_people:
                final_assignments.append({
                    'id': person['id'],
                    'name': person['name'],
                    'condition': condition,
                    'room': room
                })

    return final_assignments

# Perform the room assignment ensuring the capacity constraint is met for each condition
people_correctly_assigned = assign_rooms_properly(people, room_capacities)

# Creating the final DataFrame and saving it to CSV
df_correct_final = pd.DataFrame(people_correctly_assigned, columns=['id', 'name', 'room', 'condition'])
df_correct_final.to_csv('people_assignments_correct_final.csv', index=False)
df_correct_final.head()
```
