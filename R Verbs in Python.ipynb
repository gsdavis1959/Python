```python
import numpy as np
import pandas as pd
from siuba import _
from siuba.dply.verbs import group_by, mutate, select, summarize, ungroup
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gsdavis1959/Data/master/HR-Employee-Attrition.csv')
df.head()
```

```python
df_attrition = df >> \
    group_by('Attrition') >> \
    summarize(
            mean_daily_rate = np.mean(_.DailyRate))
```
