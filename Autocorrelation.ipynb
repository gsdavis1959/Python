```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import acf
```

```python
df = pd.DataFrame({'A':[1,2,3,4,5,], 'B': [2,6,8, 12, 22]})
df
```

```python
# Calculate the autocorrelation of each variable
acf_df = pd.DataFrame()
for col in df.columns:
    acf_df[col] = acf(df[col], nlags=10)

# Plot the autocorrelation of each variable
acf_df.plot()
```

```python
import pandas as pd
import numpy as np

np.random.seed(10)
df = pd.DataFrame(np.random.randn(10,5))
corr = df.corr()
corr_pairs = []
for i in range(len(corr.columns)):
    for j in range(i+1, len(corr.columns)):
        corr_pairs.append((corr.columns[i], corr.columns[j], corr.iloc[i, j]))
corr_pairs = sorted(corr_pairs, key=lambda x: x[2], reverse=True)
corr_pairs
```
